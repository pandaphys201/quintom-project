```python
import numpy as np
```

```python
# LambddaCDM
# 1. Define the standard deviations (1-sigma errors)
# For l_A, we use 0.0895 (the average of the asymmetric +0.089, -0.090 errors)
std_devs = np.array([0.0046, 0.0895, 0.00015])

# 2. Define the normalized correlation matrix (rho)
rho = np.array([
    [ 1.0,   0.46, -0.66],
    [ 0.46,  1.0,  -0.33],
    [-0.66, -0.33,  1.0]
])

# 3. Create the diagonal matrix D
# np.diag() places the standard deviations on the main diagonal and zeros everywhere else
D = np.diag(std_devs)

# 4. Calculate the unnormalized covariance matrix (Sigma)
# The '@' operator in Python is used for matrix multiplication
covariance_matrix = D @ rho @ D

# 5. Calculate the Inverse Covariance Matrix (Precision Matrix)
inverse_covariance_matrix = np.linalg.inv(covariance_matrix)

# Display the results with formatted output for readability
print("Unnormalized Covariance Matrix:")
np.set_printoptions(suppress=True, precision=8) # Keeps numbers from printing in scientific notation if possible
print(covariance_matrix)

# Display the results
print("Inverse Covariance Matrix:")
np.set_printoptions(suppress=True, precision=2)
print(inverse_covariance_matrix)
```
